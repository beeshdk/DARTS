```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
from math import sqrt,e 
from math import log,pi
import timeit

# -- coding: utf-8 --
"""
Created on Fri Oct 30 13:05:09 2020

@author: soffi
"""
    
import pandas as pd

#PRW_4 (80)
results = pd.read_excel (r'Results_BASE(mid,0.5,80)_PRW4.xlsx')
time = pd.DataFrame(results, columns= ['time'])
temp_out_PRW_04 = pd.DataFrame(results, columns= ['PRW_04 : temperature (K)'])
energy_PRW_04 = pd.DataFrame(results, columns= ['PRW_04 : energy (KJ/day)'])
water_PRW_04 = pd.DataFrame(results, columns= ['PRW_04 : water rate (m3/day)'])

#PRW_4 (85)
results = pd.read_excel (r'Results_BASE(mid,0.5,85)long run_PRW4.xlsx')
time = pd.DataFrame(results, columns= ['time'])
temp_out_PRW_04_85 = pd.DataFrame(results, columns= ['PRW_04 : temperature (K)'])
energy_PRW_04_85 = pd.DataFrame(results, columns= ['PRW_04 : energy (KJ/day)'])
water_PRW_04_85 = pd.DataFrame(results, columns= ['PRW_04 : water rate (m3/day)'])

plt.figure()
plt.plot(time, temp_out_PRW_04, label='80')
plt.plot(time, temp_out_PRW_04_85, label='85')
plt.xlabel("Time [days]")
plt.ylabel("Temperature Output [K]")
plt.title('Temperature Output at PRW_04', fontsize=20)
plt.legend()

plt.figure()
plt.plot(time, -1*energy_PRW_04, label='80')
plt.plot(time, -1*energy_PRW_04_85, label='85')
plt.xlabel("Time [days]")
plt.ylabel("Energy Output [KJ/day]")
plt.title('Energy Output at PRW_04', fontsize=20)
plt.legend()

plt.figure()
plt.plot(time, -1*water_PRW_04, label='80')
plt.plot(time, -1*water_PRW_04_85, label='85')
plt.xlabel("Time [days]")
plt.ylabel("Flow rate [m3/day]")
plt.title('Flow Rate at PRW_04', fontsize=20)
plt.legend()
```
